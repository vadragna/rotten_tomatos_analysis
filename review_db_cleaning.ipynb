I need to standardize this value. So, I will first convert the NaNs into 0 so I will remember them and then get the ratio and multiply per 10

```python
def convert_strings(x):
    count = 0
    try:
        if isinstance(x, (int, float)):
            return x
        elif isinstance(x, str):
            if '/' in x:
                parts = x.split('/')
                numerator = float(parts[0])
                denominator = float(parts[1])
                score = (numerator / denominator) * 100
                return score
            elif 'out of' in x:
                parts = x.split(' out of ')
                numerator = float(parts[0])
                denominator = float(parts[1])
                score = (numerator / denominator) * 100
                return score
            else:
                return 'Do not know what to do'
    except:
        count =+ 1
        print('could not convert', x, count)
```

```python
x = convert_strings('6 out of 7')
x
```

```python
reviews['originalScore'] = reviews['originalScore'].apply(convert_strings)
```

I am going to convert all string to float 0.0, so I will keep in mind they are non-desumible values

```python
reviews['originalScore'] = pd.to_numeric(reviews['originalScore'], errors='coerce').fillna(0.0)
```

```python
reviews['originalScore'].max()
```

```python
count_out_of_range = ((reviews['originalScore'] > 100) | (reviews['originalScore'] < 0)).sum()
count_out_of_range
```

```python
reviews['originalScore'] = reviews['originalScore'].apply(lambda x: 0 if x > 100 or x < 0 else x)
```
