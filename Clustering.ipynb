```python
positive_reviews = reviews[reviews['reviewState'] == 'fresh']
```

```python
positive_reviews.shape
```

```python
def list_critic_names(names):
    return ', '.join(names)

grouped_positive_reviews = positive_reviews.groupby('id').agg({
    'criticName': list_critic_names,  
    'originalScore': 'mean' 
}).reset_index()
```

```python
grouped_positive_reviews.head(50)
```

```python
grouped_reviews.head()
```

```python
def from_id_get_other_films_reviews_by_other_authors(x):
    movie_id_count = {}  
    subset = grouped_positive_reviews[grouped_positive_reviews['id'] == x]
    
    for authors in subset['criticName']:
        parts = authors.split(',')
        
        for author in parts:
            subset2 = grouped_positive_reviews[grouped_positive_reviews['criticName'].str.contains(author)]           
            subset2 = subset2[subset2['id'] != x]
            
            for movie_id in subset2['id']:
                if movie_id in movie_id_count:
                    movie_id_count[movie_id] += 1
                else:
                    movie_id_count[movie_id] = 1
    
    sorted_items = sorted(movie_id_count.items(), key=lambda x: x[1], reverse=True)

    top_ten_items = sorted_items[:10]
               
    return top_ten_items
```

```python
# Call the function to get the count of movie titles and the list of other reviewed movies
title_count = from_id_get_other_films_reviews_by_other_authors('mission_impossible')

# Print the movie title count
print(title_count)
```

```python
def get_id_from_input(x):
    x = x.lower()
    
    base_url = 'http://www.omdbapi.com/'

    params = {
    't': x,  
    'apikey': api_key,               
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        movie_data = response.json()  
        print('Searching for similiar movies to ', movie_data['Title'], ' from ', movie_data['Year'], ' by ', movie_data['Director'])
    else:
        print('API request failed with status code:', response.status_code)
      
    matching_titles = films[(films['title'].str.lower() == x) & (films['director'].str.lower() == movie_data['Director'].lower())]

    if not matching_titles.empty:
        return matching_titles.iloc[0]['id']
    else:
        return None
```

```python
def get_title_from_id(x):
    title = films[films['id'] == x].values[0][0]
    return title
```

```python
def better_recommender():
    # other_reviews_recommender = []
    user_input = input('Tell me a movie you like...')
    movie_id = get_id_from_input(user_input)
    cluster = get_cluster_from_input(user_input)
    other_reviews_recommender = from_id_get_other_films_reviews_by_other_authors(movie_id)
    # print('other_reviews_recommender', other_reviews_recommender)
    for other_reviews in other_reviews_recommender:
        movie_id = other_reviews[0]
        title = get_title_from_id(movie_id)
        print('title', title)
        matching_movies = films.loc[films['id'] == movie_id, 'clusters']       
        if not matching_movies.empty and matching_movies.values[0] == cluster:
            print('Found a movie according to other reviewers:', title)
            return other_reviews
    print('getting a recommendation from cluster')
    return get_recommendation_in_cluster(cluster)
    print('movie_id', movie_id, 'cluster', cluster)
```
